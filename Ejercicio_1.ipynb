crea la funcion average_list() que recive una lista y devuleve otra lista de la misma longitud, cuyo elementos son el promedio del original y sus vecinos
Es un *conversor* de toda la vida, mas viejo que el TBO.

*La operación elemental requiere del elemento y otros más (los vecinos).
*No todo elemento tiene los mismos vecinos.

Elimina los casos especiales.

### Técnica del centinela: eliminas el primero y el útimo, con el None.

Ahora todo elemento tiene los mismos vecinos: 2.

### ¿Cómo averiguo qué vecinos tiene un elemento?

* Tiene que iterar por índices, porque el índice te da el vecino
*i -1, i + 1 (vecinos)

## El problema original *desestructurado*

*interar por una secuencia construyendo otra (conversor)
*Encontrar los vecinos (indices) de cada elemento: una lista de indices de los vecinos y mi indice.
*Sacar un promedico de esos tres números (el elemento indicado por cada indice).


```python
num =[float | int]

def average_list(numbers: list[num])->list[num]:

  averaged = []

  #Al rescate centinela
  preprocessed = [None] + numbers + [None]

  #iterar con enumerate los numeros
  for item, current in enumerate(preprocessed):

  # Omitir none
   if current is None:
    continue

  # Llamamos la funcion def get_indexes
   indices = get_indexes(item)

  # Llamamos la funcion get_values
   values = get_values(indices, preprocessed)

  # Se crea una lista y se ignora el none
   values_sum = [valor for valor in values if valor is not None]

  # Se evulua el promedio
   promedio = sum(values_sum) / len(values_sum)
   averaged.append(promedio)

  return averaged

```

```python
def get_indexes(original_index: int)->list[int]:
  # Me entrega 3 posiciones atra, actual, adealnte.
  return [original_index - 1, original_index, original_index + 1]
```

```python
def get_values(indices: list[int], seq: list[None|float|int])->list[float|int]:
  # Busca en la lista el valor en cada una.
  return [seq[item] for item in indices]
```

```python
datos = [ 1, 2, 3]
resultado = average_list(datos)
print(resultado)
```
